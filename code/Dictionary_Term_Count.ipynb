```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict
```

Aggregate ngram files are generated by `ngram_agg.py`:

`python3 ngram_agg.py <path-to-jstor-directory> <which-ngram> <path-to-filtered-article-name-file> <path-to-output-file>`

The script will combine the counts for all articles whose names are in `<path-to-filtered-article-name-file>` for the ngram `<which-ngram>` into a single `.txt` file `<path-to-output-file>` having the same form as any raw ngram file.

```python
with open('ngram1_agg.txt', 'r') as f:
    ngram1 = dict((x.split()[0], int(x.split()[1])) for x in f.read().split('\n')[:-1])
```

```python
with open('ngram2_agg.txt', 'r') as f:
    ngram2 = dict((' '.join(x.split()[:2]), int(x.split()[2])) for x in f.read().split('\n')[:-1])
```

```python
with open('ngram3_agg.txt', 'r') as f:
    ngram3 = dict((' '.join(x.split()[:3]), int(x.split()[3])) for x in f.read().split('\n')[:-1])
```

```python
ngram = ngram3
ngram.update(ngram2)
ngram.update(ngram1)
```

```python
def term_count(file, ngram):
    with open(file, 'r') as f:
        tc = dict((x.replace(',', ' '), ngram[x.replace(',', ' ')]) if x.replace(',', ' ') in ngram else (x.replace(',', ' '), 0) for x in f.read().split('\n')[:-1])
        return OrderedDict(sorted(tc.items(), key=lambda kv: kv[1], reverse=True))
```

```python
ngram_values = list(ngram.values())
```

```python
def tc_plot(file):
    tc = term_count(file, ngram)
    print('\n'.join(['\t'.join((k, str(v))) for k, v in tc.items()]))
    tc_values = list(tc.values())
    plt.figure(figsize=(15, 5))
    plt.xlim((0, max(tc_values) * 1.01))
    sns.distplot(ngram_values, hist=False)
    sns.scatterplot(x=tc_values, y=0)
    plt.xlabel('Term Count')
    plt.ylabel('Frequency')
    plt.title('Distribution of Term Counts')
```

```python
tc_plot('../Dictionaries/Cultural_core.csv')
```
